```python
# ===============================================
# Liebherr Hackathon 2025 – Root Cause Analysis
# ===============================================
# Teamname: <Euer Teamname>
# Mitglieder: <Namen>
# ===============================================

# --- 0. Setup ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.tseries.offsets import CustomBusinessHour
from datetime import datetime

# Optional: Darstellungsoptionen
pd.set_option("display.max_columns", None)
sns.set(style="whitegrid")

# --- 1. Daten einlesen ---
# Beispiel:
df = pd.read_csv("2025-10-22-qualification_data.csv")
# oder:
# df = pd.read_excel("fertigungsdaten.xlsx")
```

```python

# --- 3. Datentypen & Bereinigungen ---
time_cols = [
    "AFO_Start_SOLL", "AFO_Ende_SOLL",
    "AFO_Start_IST", "AFO_Ende_IST",
    "Auftragseingang", "Auftragsende_SOLL"
]
for col in time_cols:
    df[col] = pd.to_datetime(df[col], errors="coerce")

# Fehlerhafte oder fehlende Zeiten ggf. filtern
df = df.dropna(subset=["AFO_Start_IST", "AFO_Ende_IST"])
```

```python

# --- 4. Feature Engineering ---
df["AFO_Dauer_SOLL"] = (df["AFO_Ende_SOLL"] - df["AFO_Start_SOLL"]).dt.total_seconds() / 3600
df["AFO_Dauer_IST"] = (df["AFO_Ende_IST"] - df["AFO_Start_IST"]).dt.total_seconds() / 3600
df["AFO_Verspätung_h"] = (df["AFO_Ende_IST"] - df["AFO_Ende_SOLL"]).dt.total_seconds() / 3600

# Verspätung pro Auftrag (Ende des letzten AFO vs. Soll-Ende)
auftrag_delay = (
    df.groupby("AuftragsID")["AFO_Ende_IST"].max() - df.groupby("AuftragsID")["Auftragsende_SOLL"].first()
).dt.total_seconds() / 3600
df_delay = pd.DataFrame({"AuftragsID": auftrag_delay.index, "Auftrag_Verspätung_h": auftrag_delay.values})
```

```python

# --- 5. Grundlegende Statistik ---
print("Durchschnittliche AFO-Verspätung (h):", df["AFO_Verspätung_h"].mean())
print("Durchschnittliche Auftragsverspätung (h):", df_delay["Auftrag_Verspätung_h"].mean())
```

```python

# --- 6. Einflussfaktoren prüfen ---

# 6.1 Maschinenbezeichnung
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="Maschinenbezeichnung", y="AFO_Verspätung_h")
plt.title("Verteilung der AFO-Verspätung pro Maschinentyp")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6.2 Bauteil
plt.figure(figsize=(10, 5))
sns.boxplot(data=df, x="Bauteilbezeichnung", y="AFO_Verspätung_h")
plt.title("AFO-Verspätung nach Bauteil")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 6.3 Priorität
sns.barplot(data=df, x="Priorität", y="AFO_Verspätung_h", estimator=np.mean)
plt.title("Durchschnittliche AFO-Verspätung nach Priorität")
plt.show()

# 6.4 Wochentagseffekt
df["Wochentag"] = df["AFO_Start_IST"].dt.day_name()
sns.barplot(data=df, x="Wochentag", y="AFO_Verspätung_h", estimator=np.mean,
            order=["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"])
plt.title("Durchschnittliche AFO-Verspätung pro Wochentag")
plt.show()
```

```python

# --- 7. Engpassanalyse: Arbeitsschritte ---
top_afo = df.groupby("Arbeitsschrittbezeichnung")["AFO_Verspätung_h"].mean().sort_values(ascending=False)
print("Top 10 verspätete AFOs:\n", top_afo.head(10))
```

```python

# --- 8. (Optional) Pareto-Analyse ---
top_causes = top_afo.cumsum() / top_afo.sum()
plt.figure(figsize=(8,4))
top_afo.head(20).plot(kind="bar")
plt.title("Pareto-Analyse: häufigste Verspätungs-Ursachen (AFO)")
plt.ylabel("Ø Verspätung [h]")
plt.show()
```
